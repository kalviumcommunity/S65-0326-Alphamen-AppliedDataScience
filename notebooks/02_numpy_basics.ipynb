# NumPy Basics

## Creating arrays from Python lists

NumPy arrays make numerical analysis faster and more consistent than plain Python lists for many data tasks.

```python
import numpy as np

complaint_counts = [12, 18, 15, 21]
complaint_array = np.array(complaint_counts)
complaint_array
```

## Shape, dimensions, and index positions

These properties help describe how data is laid out in memory and how to access values safely.

```python
complaint_matrix = np.array([[12, 18], [15, 21]])
complaint_array.shape, complaint_array.ndim, complaint_array[1], complaint_matrix[0, 1]
```

## Basic mathematical operations

NumPy arrays support element-wise arithmetic for fast numeric work.

```python
complaint_array + 2, complaint_array * 2, complaint_array.mean()
```

## Vectorized operations instead of Python loops

Vectorized expressions are shorter and usually faster than manual loops for array-wide calculations.

```python
adjusted_counts = complaint_array * 1.1
adjusted_counts
```
