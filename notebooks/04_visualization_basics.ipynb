# Visualization Basics

## Histograms

Histograms help show the distribution of a numeric variable such as complaint response time.

```python
import matplotlib.pyplot as plt
import pandas as pd

complaints_df = pd.read_csv("data/raw/municipal_complaints.csv")
plt.hist(complaints_df["response_time_hours"], bins=5)
plt.xlabel("Response Time (hours)")
plt.ylabel("Frequency")
plt.title("Distribution of Complaint Response Times")
plt.show()
```

## Boxplots

Boxplots help summarize spread and highlight possible extreme values.

```python
plt.boxplot(complaints_df["response_time_hours"])
plt.ylabel("Response Time (hours)")
plt.title("Complaint Response Time Boxplot")
plt.show()
```

## Line plots for trends over time

Line plots help show how values change across an ordered timeline.

```python
trend_df = complaints_df.copy()
trend_df["timestamp"] = pd.to_datetime(trend_df["timestamp"])
daily_counts = trend_df.groupby(trend_df["timestamp"].dt.date).size()
plt.plot(daily_counts.index.astype(str), daily_counts.values, marker="o")
plt.xlabel("Date")
plt.ylabel("Complaint Count")
plt.title("Daily Complaint Volume")
plt.xticks(rotation=45)
plt.show()
```

## Scatter plots for relationships between variables

Scatter plots make it easier to inspect whether two numeric variables move together.

```python
scatter_df = complaints_df.assign(description_length=complaints_df["description"].str.len())
plt.scatter(scatter_df["description_length"], scatter_df["response_time_hours"])
plt.xlabel("Description Length")
plt.ylabel("Response Time (hours)")
plt.title("Description Length vs Response Time")
plt.show()
```

## Outlier detection

Outliers can be explored visually and with a simple IQR-based rule.

```python
q1 = complaints_df["response_time_hours"].quantile(0.25)
q3 = complaints_df["response_time_hours"].quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr
complaints_df[complaints_df["response_time_hours"] > upper_bound]
```
