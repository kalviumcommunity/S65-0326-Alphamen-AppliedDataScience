## Creating Series from lists and arrays

Pandas Series add labels on top of one-dimensional data, which makes analysis more expressive.

```python
import numpy as np
import pandas as pd

complaint_list = [12, 18, 15]
complaint_array = np.array([21, 17, 19])

series_from_list = pd.Series(complaint_list)
series_from_array = pd.Series(complaint_array)
series_from_list, series_from_array
```
