# Pandas Foundations

## Creating Series from lists and arrays

Pandas Series add labels on top of one-dimensional data, which makes analysis more expressive.

```python
import numpy as np
import pandas as pd

complaint_list = [12, 18, 15]
complaint_array = np.array([21, 17, 19])

series_from_list = pd.Series(complaint_list)
series_from_array = pd.Series(complaint_array)
series_from_list, series_from_array
```

## Creating DataFrames from dictionaries and files

DataFrames are table-shaped structures that work well for real project datasets.

```python
complaint_frame = pd.DataFrame({
    "category": ["water", "roads", "garbage"],
    "count": [12, 18, 15],
})

complaint_frame
```

## Loading CSV data into a DataFrame

Most real projects start by loading tabular data from a file.

```python
complaints_df = pd.read_csv("data/raw/municipal_complaints.csv")
complaints_df
```

## Inspecting DataFrames with head, info, and describe

These helpers provide a fast first look at data contents, schema, and summary statistics.

```python
complaints_df.head()
complaints_df.info()
complaints_df.describe(include="all")
```

## Data shape and column data types

Shapes and dtypes help confirm that a dataset matches expectations before deeper analysis.

```python
complaints_df.shape, complaints_df.dtypes
```

## Selecting rows and columns

Indexing and slicing make it easier to focus on specific pieces of a dataset.

```python
complaints_df.loc[0:2, ["category", "status"]]
complaints_df.iloc[:3, :3]
```

## Detecting missing values

Missing value checks are part of early data quality assessment.

```python
quality_df = complaints_df.copy()
quality_df.loc[2, "status"] = None
quality_df.isna().sum()
```

## Handling missing values

Drop and fill strategies depend on whether missing data can be safely replaced.

```python
quality_df.fillna({"status": "unknown"})
quality_df.dropna()
```

## Identifying and removing duplicates

Duplicate checks help avoid overstating complaint counts.

```python
duplicate_df = pd.concat([complaints_df, complaints_df.iloc[[0]]], ignore_index=True)
duplicate_df.duplicated().sum(), duplicate_df.drop_duplicates().shape
```

## Standardizing column names and data formats

Consistent names and formats reduce friction in later analysis.

```python
standardized_df = complaints_df.copy()
standardized_df.columns = standardized_df.columns.str.lower()
standardized_df["timestamp"] = pd.to_datetime(standardized_df["timestamp"])
standardized_df.dtypes
```

## Summary statistics for individual columns

Column-level summaries provide a fast sense of scale and spread.

```python
complaints_df["response_time_hours"].describe()
```
