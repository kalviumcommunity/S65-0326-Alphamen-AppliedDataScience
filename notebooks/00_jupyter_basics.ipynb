## Running and managing kernels

- Run a cell with Shift + Enter.
- Restart the kernel when variables get out of sync or after package changes.
- Interrupt the kernel if a cell is stuck in a long loop or waiting on a slow operation.
- Re-run important cells after a restart because notebook memory resets.
