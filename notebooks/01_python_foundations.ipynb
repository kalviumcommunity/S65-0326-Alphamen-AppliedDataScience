# Python Foundations

## Numeric and string data types

Integers and floats help describe counts, averages, and rates. Strings are useful for complaint categories, status values, and locations.

```python
complaint_count = 42
average_delay_hours = 3.5
complaint_category = "water supply"
complaint_status = "pending"

print(type(complaint_count))
print(type(average_delay_hours))
print(type(complaint_category))
print(type(complaint_status))
```

## Lists, tuples, and dictionaries

Lists are useful for ordered values, tuples for fixed grouped values, and dictionaries for named fields in a record.

```python
daily_complaints = [12, 18, 15, 21]
ward_location = (18.52, 73.85)
complaint_record = {
    "category": "road damage",
    "status": "resolved",
    "ward": "A1",
}

daily_complaints, ward_location, complaint_record
```

## Conditional statements

Conditionals help express business rules such as whether a complaint needs escalation.

```python
response_hours = 30

if response_hours > 24:
    escalation_message = "Escalate this complaint"
else:
    escalation_message = "Within expected response time"

escalation_message
```

## Iteration with for and while loops

Loops are useful when checking repeated records such as daily complaint volumes.

```python
daily_complaints = [12, 18, 15]
for count in daily_complaints:
    print(count)

index = 0
while index < len(daily_complaints):
    print(daily_complaints[index])
    index += 1
```

## Defining and calling functions

Functions help package repeated logic into one named unit of work.

```python
def summarize_total(values):
    return sum(values)

summarize_total([12, 18, 15])
```

## Passing data into functions and returning results

Inputs make functions flexible, and return values make them reusable in later analysis steps.

```python
def average_complaints(values):
    return sum(values) / len(values)

average_complaints([12, 18, 15])
```
